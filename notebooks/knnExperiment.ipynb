### Import the KNN library which you implemented from scratch

```python
from knn.knn import KNearestNeighbour
```

### Import other essential libraries

```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
colourMap = ListedColormap(["#FF0000", "00FF00", "000FF"])
```

### Load the dataset

```python
irisDataset = datasets.load_iris()
X, y = irisDataset["data"], irisDataset["target"]
```

### Split into training and testing data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=34)
```

```python
plt.figure()
plt.scatter(X[:, 2], X[:, 3], c=y, edgecolors='k', s=20)
plt.show()
```

### Define the model for training

```python
classifier = KNearestNeighbour(k=5)
classifier.fit(X_train, y_train)
pred = classifier.predict(X_test)
accuracy = KNearestNeighbour.accuracyScore(pred, y_test)
accuracy
```
